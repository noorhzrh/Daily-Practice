```python
# x = np.array([-2,-1.5,-1,0,1,1.5,2])
x = np.linspace(-15,5,100) # 100 refers to the nnumber of edges/steps/stops 
e = 2*((x+3)**2)+3

result = np.array(e)

plt.plot(x,result)
plt.axvline(0, color = "black", linewidth = 0.4)
plt.axhline(0, color = "black", linewidth = 0.4)
```

```python
# x = np.array([-2,-1.5,-1,0,1,1.5,2])
x = np.linspace(-10000,10000,10000) # 100 refers to the nnumber of edges/steps/stops 
e = 3*(x**3)+(2*(x**2))+((x-1)**2)+4

result = np.array(e)

plt.plot(x,result)
plt.axvline(0, color = "black", linewidth = 0.4)
plt.axhline(0, color = "black", linewidth = 0.4)
```
