```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.DataFrame(data)
df
```

```python
sns.histplot(data = df, x= 'height', bins=3, kde=True) # bins is like partition of the bars
```

```python
malay = df[df['ethnicity'] == 'malay']
print(malay)
```

```python
female = df[df['gender'] == 'f']
print(female)
```

```python
malayOver21 = df[   (df['ethnicity'] == 'malay') & (df['age'] >= 21)] 
print(malayOver21)
```
