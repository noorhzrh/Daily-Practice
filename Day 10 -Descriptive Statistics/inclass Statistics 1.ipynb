```python
nums = [20,89,130,50,90,80]
nums.sort()
print('sorted list nums: ',nums)
#if the number is odd,length/2
#3/2 = 1.5

mean = sum(nums)/5
print(f'print mean of the number{mean}')
if len(nums) %2 ==0:
    #if length is even 2nd position + 3rd position
    median = (nums[len(nums)//2] + nums[len(nums)//2 - 1] ) /2
    print(median)
else:
#if the length is odd
    median = nums[len(nums)//2]
    print(median)

```

```python
import numpy as np
class1 = np.array([50,47,51,49,47])
class2 = np.array([5,20,50,80,130])

mean = sum(class2)/len(class2)
print("mean is; ",mean)

#variance = sum((class2-mean)**2)/len(class2)
sumofdifference = np.array([((class2 - mean)**2)])
variance2 = (sumofdifference.sum())/len(class2)
print('variance is:',variance2)

```
