```python
import matplotlib.pyplot as plt

#create a bar plot

plt.bar(students,grades,color = 'green') # specify properties of our bar chart

plt.title('student names')

plt.xlabel('student names')

plt.ylabel('marks')

plt.show() 
```

```python
#create a bar plot

plt.scatter(students,grades)
```

```python
# line plot/line chart
plt.plot(students,grades)
```

```python
# all of them in the same place

# create a figure of 1 row 3 colums # use () cause its tuple
plt.figure( figsize=(15,5)) #define the aspect ratio #width,heightm,position

#sub plot

plt.subplot(1,3,1)
plt.bar(students,grades,color = 'green')

plt.subplot(1,3,2)
plt.plot(students,grades,color = 'green')

plt.subplot(1,3,3)
plt.scatter(students,grades,color = 'green')

```

```python
plt.figure(figsize=(15,5))
plt.title('Students-grade')
#sub plot

plt.subplot(1,3,1)
plt.bar(cities, temp, color='indigo')

plt.subplot(1,3,2)
plt.plot(cities, temp, color='pink')

plt.subplot(1,3,3)
plt.scatter(cities, temp, color='red')
```
