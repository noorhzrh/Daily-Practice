```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
crimes = pd.read_csv('C:\\Users\\user\\Desktop\\aiml\\Daily-Practice\\Day 9 - Practice Data Analysis\\crime_district.csv')
```

```python
crimes.info()
```

```python
crimes.shape
```

```python
crimes = crimes[crimes['state'] != 'Malaysia']
crimes = crimes[crimes['type'] !='all']
# titanic = titanic[titanic['Sex']!= 'male']
crimes
```

```python
crimes['date'] = pd.to_datetime(crimes['date']).dt.year
crimes['crimes_amount'] = crimes['crimes']
crimes
```

```python
plt.figure(figsize=(16, 8))
sns.barplot(data = crimes, x='state', y='crimes_amount', hue='category')
```

```python
crimes.head()
```

```python
plt.figure(figsize=(32, 8))
sns.barplot(data = crimes['state'], x='type', y='crimes_amount',hue='category')
```

```python
plt.figure(figsize=(16, 8))
sns.barplot(data=crimes, x='year', y='crimes_amount')
```
