```python

import seaborn as sns

titanic=sns.load_dataset('titanic')
titanic.shape # to check how much rows and columns
```

```python
titanic = titanic.dropna() #derop anhthing that has missing value/NaN
```

```python
titanic['age'] = titanic['age'].astype(int)

titanic
```

```python
titanic['fare'] = titanic['age'].round(2)
```

```python
sns.barplot(data=titanic, x='sex', y='survived',color='red')
```

```python
sns.pointplot(data=titanic, x='age', y='class')
```

```python
sns.pairplot(titanic)
```
