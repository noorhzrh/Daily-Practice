```python
# create a new df but without the species column
#             (target the element,direction of target)
new_df=df.drop("species",axis=1)
### or ###
#new_df=df.drop(columns="species")
```

```python


# kalau nak targetbanyk column pakai list []
sepalarea = new_df[['sepal_length','sepal_width']]
sepalarea.head()
```

```python
#calculated column
# create variable to store first, then do the calculation
sepalarea = new_df['sepal_length'] * new_df['sepal_width']
           #df,target object, calc, df,target obj
sepalarea.head()
```

```python
#display the sepal area as a new column in the new_df 
new_df['sepal_area'] = sepalarea

new_df.head()
```

```python
#for petal area
petalarea = new_df['petal_length'] * new_df['petal_width']
petalarea.head()
```

```python
new_df['petal_area']=petalarea
new_df.head(50)
```

```python
#filter out parts of dataframe using pandas

#only part of iris without_species value of sepal_area is >14

largeFlower = new_df[new_df['sepal_area'] >15 ] 
largeFlower.head(10)

largeFlower.info()
```

```python
largeFlower.to_csv('largeFlower.csv')
```

```python
largeFlower.head()
```

```python
#how many flowers have a sepal area exactly 16
Flower16 = new_df[new_df['sepalarea']==16]

Flower16.info()
```
