```python
# dictionary

#create a small dictinary 
#store names of 4 students and their marks

student_marks = {
    #key : value
    'Ahmad':78,
    'Siti':88,
    'Bob':85,
    'John':90
    }


print(student_marks)

# to access value by key

print(student_marks['Ahmad'])

for student, marks in student_marks.items():
    print(f"{student}:{marks}")
```
