```python
columns = ['total_bill','tip','size'] # define colums needed 
covariance2 = tipsdf[columns].cov() # cpovariance = df[columns].cov()
covariance2.round(2) 
```

### Correlation ###

```python
correlation2 = tipsdf[columns].corr()
correlation2.round(2)
```

```python
titanicdf = sns.load_dataset('titanic')
titanicdf
```

```python
columns = ['age','fare','pclass','survived','parch','sibsp']
covariance = titanicdf[columns].cov()
covariance.round(2)
```

```python
columns = ['age','fare','pclass','survived','parch','sibsp']
correlation = titanicdf[columns].corr(method= 'pearson') #defaulth method- pearson
correlation.round(2)
```

### Correlation Heat Map ###

```python
sns.heatmap(correlation2,cmap = 'Blues', annot=True,fmt='.2f', vmin = -1, vmax =1)
#coolwarm, Greens, Blue
```

```python
irisdf = sns.load_dataset('iris')
irisdf
```

```python
columns = ['sepal_length','sepal_width','petal_length','petal_width']
iriscorr = irisdf[columns].corr(method='pearson')
iriscorr.round(2)
```

```python
tipsdf['tip_pct'] =((tipsdf['tip'] / tipsdf['total_bill'])*100).round(2)
tipsdf

```

```python
columns = ['total_bill','tip','tip_pct']
tipscorr = tipsdf[columns].corr()
tipscorr.round(2)
```
